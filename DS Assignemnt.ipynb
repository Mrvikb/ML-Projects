```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (12.0, 9.0)
```

```python
X = pd.read_csv("linearX.csv", header=None)
```

```python
X.columns = ['x']
```

```python
Y = pd.read_csv("linearY.csv", header=None)
Y.columns = ['y']
```

```python
X1 = X.iloc[:,0]
Y1 = Y.iloc[:,0]
plt.scatter(X, Y)
plt.show()
```

# Implementing Gradient Descent 

```python
m = 0.2
c = 0.2

L = 0.0001  # The learning Rate
epochs = 1000  # The number of iterations to perform gradient descent

n = float(len(X1)) # Number of elements in X

# Performing Gradient Descent 
for i in range(epochs): 
    Y_pred = m*X1 + c  # The current predicted value of Y
    D_m = (-1/n) * sum(X1 * (Y1 - Y_pred))  # Derivative wrt m
    D_c = (-1/n) * sum(Y1 - Y_pred)  # Derivative wrt c
    m = m - L * D_m  # Update m
    c = c - L * D_c  # Update c
    
print (m, c)
```

```python
# Making predictions
Y_pred = m*X1 + c

#plt.scatter(X1, Y1) 
plt.scatter(X1, Y1, label = "Actual")
plt.scatter(X1, Y_pred, label = "Predicted")
plt.xlabel('Acidity')
plt.ylabel('Density')
plt.title('Comparison of Actual vs Predicted')
#plt.plot([min(X1), max(X1)], [min(Y_pred), max(Y_pred)], color='red')  # regression line
plt.legend()
plt.show()
```
